1. 2주 간의 판매 데이터를 하나의 DataFrame으로 결합하세요. week1 DataFrame에 'week1'이라는 키를 할당하고 week2 DataFrame에 'week2' 라는 키를 할당하세요

```python
pd.concat(objs = [week1, week2], keys = ['week1', 'week2'])
```

2. 2주동안 매주 식당에서 식사한 고객을 찾으세요.

```python
week1.merge(right = week2, how = 'inner', on = 'Customer ID').head()
```

```python
week1.merge(right = week2, how = 'inner', on = 'Customer ID').drop_duplicates(subset = ['Customer ID']).head()
```

3. 2주동안 매주 식당에서 식사를 하고 같은 음식을 주문한 고객을 찾으세요

```python
week1.merge(right = week2, how = 'inner', on = ['Customer ID', 'Food ID'])
```

4. 1주차에만 방문한 고객과 2주차에만 방문한 고객이 누구인지 확인하세요

```python
week1.merge(right = week2, how = 'outer', on = 'Customer ID', indicator = True)
```

5. week1 DataFrame의 각 행은 음식을 주문한 고객을 나타냅니다. 각 행에 대한 고객의 정보를 customers DataFrame에서 가져오세요

```python
week1.merge(right = customers, how = 'left', left_on = 'Customer ID', right_index = True).head()
```
