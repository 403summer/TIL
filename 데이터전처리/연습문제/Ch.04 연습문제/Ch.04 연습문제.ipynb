```python
# 01. nfl.csv 파일을 가져오려면 어떻게 해야 할까요? Birthday 열의 값을 날짜/시간으로 변환하는 효과적인 방법은 무엇일까요?

import pandas as pd

nfl = pd.read_csv('nfl.csv', parse_dates = ['Birthday'])
nfl
```

```python
# 02. 선수의 이름을 DataFrame의 인덱스로 지정하는 두 가지 방법은 무엇인가요?

# 다음 두 줄은 결과가 동일합니다.
pd.read_csv('nfl.csv', index_col= 'Name', parse_dates = ['Birthday'])
nfl.set_index('Name')
```

```python
# 03. 이 데이터셋에서 팀 당 선수가 몇 명인지 계산하는 방법은 무엇인가요?

# 다음 두 줄은 결과가 동일합니다.
nfl.Team.value_counts().head()
nfl['Team'].value_counts().head()
```

```python
# 04. 가장 높은 연봉을 받는 5명의 선수는 누구인가요?

nfl.sort_values('Salary', ascending = False).head()
```

```python
# 05. 데이터셋에서 먼저 팀을 알파벳 순서로 정렬한 다음 연봉을 내림차순으로 정렬하려면 어떻게 해야하나요?

nfl.sort_values(['Team', 'Salary'], ascending = [True, False])
```

```python
# 06. New York Jets 팀에서 가장 나이가 많은 선수는 누구이며 그 의 생일은 언제인가요?

nfl = nfl.set_index(keys = 'Team')
nfl.head()
```
