```python
import pandas as pd

investments = pd.read_csv('investments.csv')
investments.head()
```

```python
investments.nunique()
```

```python
investments = investments.set_index(
    keys = ['Status', 'Funding Rounds', 'State']
).sort_index()
```

```python
investments.head()
```

1. status가 'Closed'인 모든 행을 추출하세요

```python
investments.loc[('Closed',)].head()
```

2. Status가 'Acquired' 이고 Funding Rounds가 10인 모든 행을 추출하세요

```python
investments.loc[('Acquired', 10)]
```

3. Status가 'Operating'이고 Funding Round가 6이고 State가 'NJ'인 모든 행을 추출하세요

```python
investments.loc[('Operating', 6, 'NJ')]
```

4. Status가 'Closed'이고 Funding Rounds가 8인 모든 행을 추출하세요. 이때 Name 열만 추출하세요

```python
investments.loc[('Closed', 8), ('Name',)]
```

5. Status가 Funding Rounds의 값과 상관없이 Status가 'NJ'인 모든 행을 추출하세요

```python
# 다음 두 줄은 결과가 동일합니다
investments.xs(key = 'NJ', level = 2).head()
investments.xs(key = 'NJ', level = 'State').head()
```
