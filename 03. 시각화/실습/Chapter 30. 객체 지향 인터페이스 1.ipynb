# plt.subplots() 사용하기

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
plt.show()
```

```python
# fig, ax = plt.subplots()
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

plt.show()
```

# 행과 열 설정하기 (nrows, ncols)

```python
fig, ax = plt.subplots(2, 2)
plt.show()
```

# X, Y축 공유하기 (sharex, sharey)

```python
fig, ax = plt.subplots(2, 2, sharex = True, sharey = True)
plt.show()
```
