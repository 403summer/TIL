# 기본 사용

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'r--', x, x**2, 'bo', x, x**3, 'g-.')
plt.show()
```

# 스타일 지정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color = '#e35f62', marker = '*', linewidth = 2)
plt.plot(x, x**3, color = 'forestgreen', marker = '^', markersize = 9)
plt.show()
```
