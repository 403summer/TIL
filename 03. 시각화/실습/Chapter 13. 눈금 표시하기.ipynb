# 기본 사용

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color = '#e35f62', marker = '*', linewidth = 2)
plt.plot(x, x**3, color = 'forestgreen', marker = '^', markersize = 9)
plt.xticks([0, 1, 2])
plt.yticks(np.arange(1, 6))

plt.show()
```

# 눈금 레이블 지정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color = '#e35f62', marker = '*', linewidth = 2)
plt.plot(x, x**3, color = 'springgreen', marker = '^', markersize = 9)
plt.xticks(np.arange(0, 2, 0.2), labels = ['Jan', '', 'Feb', '', 'Mar', '',
                                          'May', '', 'June', ''])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))

plt.show()
```

# 눈금 스타일 설정하기

```python
x = np.arange(0, 2, 0.2)

plt.plot(x, x, 'bo')
plt.plot(x, x**2, color = '#e35f62', marker = '*', linewidth = 2)
plt.plot(x, x**3, color = 'springgreen', marker = '^', markersize = 9)
plt.xticks(np.arange(0, 2, 0.2), labels = ['Jan', '', 'Feb', '', 'Mar', '',
                                          'May', '', 'June', ''])
plt.yticks(np.arange(0, 7), ('0', '1GB', '2GB', '3GB', '4GB', '5GB', '6GB'))
plt.tick_params(axis = 'x', direction = 'in', length = 3, pad = 6, labelsize = 14,
              labelcolor = 'green', top = True)
plt.tick_params(axis = 'y', direction = 'inout', length = 10, pad = 15, labelsize = 12,
               width = 2, color = 'r')

plt.show()
```
