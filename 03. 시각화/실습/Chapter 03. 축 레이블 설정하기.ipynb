# 기본 사용

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.xlabel('X-Label')
plt.ylabel('Y-Label')
plt.show()
```

# 여백 지정하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Label', labelpad = 15)
plt.ylabel('Y-Label', labelpad = 20)
plt.show()
```

# 폰트 설정하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad = 15, fontdict = {'family': 'serif', 'color': 'b',
                                               'weight': 'bold', 'size': 14})
plt.ylabel('Y-Axis', labelpad = 20, fontdict = {'family': 'fantasy',
                                               'color': 'deeppink', 'weight': 'normal',
                                               'size': 'xx-large'})
plt.show()
```

```python
font1 = {'family': 'serif',
        'color': 'b',
        'weight': 'bold',
        'size': 14
       }

font2 = {'family': 'fantasy',
         'color': 'deeppink',
         'weight': 'normal',
         'size': 'xx-large'
        }

plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', labelpad = 15, fontdict = font1)
plt.ylabel('Y-Axis', labelpad = 20, fontdict = font2)
plt.show()
```

#  위치 지정하기

```python
plt.plot([1, 2, 3, 4], [2, 3, 5, 10])
plt.xlabel('X-Axis', loc = 'right')
plt.ylabel('Y-Axis', loc = 'top')
plt.show()
```
